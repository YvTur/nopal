## Analisis estadisticos Bromatologico entre cultivares

```python
##manejardor de datos
import pandas as pd
#graficador
import matplotlib.pyplot as plt
#statistical graphs
import seaborn as sns
df= pd.read_csv('tabla_1.csv').set_index('Cultivares')
df
```

```python
df.plot();
```

```python
df.plot.bar()
df.plot.bar(stacked=True)
```

```python
df.plot.hist()
df.plot.hist(alpha=0.5)
```

```python
df.plot.area(stacked=False);
```

```python
sns.heatmap(df, cmap='Blues')
```

```python
sns.heatmap(df.corr(),center=0, cmap='Greens')
```
