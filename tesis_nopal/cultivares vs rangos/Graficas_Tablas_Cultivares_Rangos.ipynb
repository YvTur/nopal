```python
##manejardor de datos
import pandas as pd
#graficador
import matplotlib.pyplot as plt
#statistical graphs
import seaborn as sns
```

## Tabla para la parte superior del Nopal

```python
df= pd.read_csv('tabla_1.csv').set_index('Cultivares')
df
```

```python
df.plot();
```

```python
df.plot.bar()
df.plot.bar(stacked=True)
```

```python
df.plot.hist()
df.plot.hist(alpha=0.5)
```

```python
df.plot.area(stacked=False);
```

```python
sns.heatmap(df, cmap='Blues')
```

```python
sns.heatmap(df.corr(),center=0, cmap='Greens')
```

```python
#En esta parte voltiamos la matriz para poder hacer la correlacion entre los cultivares
df1t = df.T
df1t
```

```python
df1t.plot();
```

```python
#En esta grafica podemos ver mayor correlacion entre los mismos cultivares(variedades) de nopal
sns.heatmap(df1t.corr(),center=0, cmap='Greens')
```
